```python
import pandas as pd
import os
import numpy as mp
```

```python

dfs = {}
path = "../Mlp/results_10_experiments/mlp"

for file in os.listdir(path):
    if file.startswith("id_") and file.endswith("_metrics.csv"):
        run_id = file[len("id_"):-len("_metrics.csv")]
        df = pd.read_csv(os.path.join(path, file))
        dfs[run_id] = df
```

```python
dfs
```

```python
len(dfs)
```

```python
resultados = []

for run_id, df in dfs.items():

    class_col = df.columns[0] 
    df[class_col] = df[class_col].astype(str).str.strip().str.lower()

    macro = df[df[class_col] == 'macro avg'].iloc[0]
    f1_macro = float(macro['f1-score'])

    acc_row = df[df[class_col] == 'accuracy'].iloc[0]
    accuracy = float(acc_row['precision'])   

    resultados.append({
        "run_id": run_id,
        "f1_macro": f1_macro,
        "accuracy": accuracy
    })

df_scores = pd.DataFrame(resultados)
```

```python
df_scores.sort_values(by="f1_macro", ascending=False).reset_index(drop=True)
```

```python
df_scores = df_scores.sort_values(by="f1_macro", ascending=False).reset_index(drop=True)
df_top5 = df_scores.head(5)

print("TOP 5:")
print(df_top5)
```

```python
mean_f1 = df_top5["f1_macro"].mean()
std_f1 = df_top5["f1_macro"].std()

mean_acc = df_top5["accuracy"].mean()
std_acc = df_top5["accuracy"].std()

print(f"Promedio F1-macro: {mean_f1:.6f}")
print(f"Std F1-macro:      {std_f1:.6f}")

print(f"Promedio Accuracy: {mean_acc:.6f}")
print(f"Std Accuracy:      {std_acc:.6f}")
```
